```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
from sklearn.datasets.samples_generator import make_blobs
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

Generate 3D Data

```
m = 200
X, y = make_blobs(n_samples=m, centers=4, n_features=3)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=X[:, 0], ys=X[:, 1], zs=X[:, 2], c=y)
```

Preprocess Data

```
mu = np.mean(X, axis=0)
s = np.std(X, axis=0)
X = (X - mu)/s
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=X[:, 0], ys=X[:, 1], zs=X[:, 2], c=y)
```

Perform PCA

```
## Calculate Covariance Matrix
cov = (1/m) * np.dot(X.T, X)

## Calculate Eigenvectors
U, S, V = np.linalg.svd(cov)

## Keep first k eigenvectors as principal components
k = 2
U_reduced = U[:, :k]
```

Calculate reduced X matrix

```
X_reduced = np.dot(X, U_reduced)
plt.scatter(x=X_reduced[:, 0], y=X_reduced[:, 1], c=y)
plt.show()
```

Determine Variance Retained in Compressed Data

```
print(f'Variance Retained: {100*sum(S[:k])/sum(S)}%')
```
